```python
import tensorflow as tf
import numpy as np
import sys
sys.path.append('../general')
from pool import Pool
```

```python
pool = Pool('../data')
train_pool, test_pool = pool.train_test_split()
```

```python
num_features = np.shape(pool.features)[1]
encode_dimention = 1024
```

```python
input_placeholder = tf.placeholder('float32', shape=(None, num_features))

encoder_weights = tf.get_variable(
    "encoder_weights", shape=(num_features, encode_dimention),
    initializer=tf.glorot_uniform_initializer()
)
encoder_bias = tf.get_variable(
    "encoder_bias", shape=(encode_dimention,),
    initializer=tf.glorot_uniform_initializer()
)

decoder_weights = tf.get_variable(
    "decoder_weights", shape=(encode_dimention, num_features),
    initializer=tf.glorot_uniform_initializer()
)
decoder_bias = tf.get_variable(
    "decoder_bias", shape=(num_features,),
    initializer=tf.glorot_uniform_initializer()
)
```

```python
encoded_features = tf.matmul(input_placeholder, encoder_weights) + encoder_bias
decoded_features = tf.matmul(encoded_features, decoder_weights) + decoder_bias
loss = tf.reduce_mean((input_placeholder - decoded_features) ** 2)
optimizer = tf.train.AdamOptimizer().minimize(
    loss,
    var_list=[encoder_weights, encoder_bias, decoder_weights, decoder_bias]
)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
train_loss = []
test_loss = []
counter = 0
```

```python
for i in range(500):
    sess.run(optimizer, {input_placeholder: train_pool.features})
    train_loss.append(sess.run(loss, {input_placeholder: train_pool.features}))
    test_loss.append(sess.run(loss, {input_placeholder: test_pool.features}))

    print('{} iter. Train: {}. Test: {}'.format(counter, train_loss[-1], test_loss[-1]))

    counter += 1
```

```python
def save(folder_name):
    variables = [encoder_weights, encoder_bias, decoder_weights, decoder_bias]
    for variable in variables:
        np.save(folder_name + '/' + variable.name, sess.run(variable))
```

```python
save('saved_weights')
```
