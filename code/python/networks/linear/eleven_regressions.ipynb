```python
import numpy as np
from eleven_regressions_model import ElevenRegressionsModel
from one_position_linear_model import OnePositionLinearModel
import sys
sys.path.append('../general')
from pool import Pool
from collections import Counter
from metric import metric
```

```python
pool = Pool('../data')
train_pool, test_pool = pool.train_test_split()
```

```python
model = ElevenRegressionsModel(len(pool.features[0]), OnePositionLinearModel)
model.teach(train_pool)
```

```python
def test_pool(model, pool):
    position_prediction = model.predict_positions(pool)
    print(metric(
        position_prediction, pool.positions,
        pool.targets, pool.probas
    ))
    print(Counter(position_prediction))
```

```python
test_pool(model, train_pool)
```
