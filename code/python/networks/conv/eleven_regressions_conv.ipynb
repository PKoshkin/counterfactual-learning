```python
import numpy as np
import sys
sys.path += ['../../general', '..']
from eleven_regressions_model import ElevenRegressionsModel
from one_position_regression_conv_model import OnePositionRegressionConvModel
from pool import Pool
from collections import Counter
from metric import metric
```

```python
pool = Pool('../../data')
train_pool, test_pool = pool.train_test_split()
```

```python
model = ElevenRegressionsModel(len(pool.features[0]), OnePositionRegressionConvModel)
model.teach(train_pool)
```

```python
def test(model, pool):
    position_prediction = model.predict_positions(pool)
    print(metric(
        position_prediction, pool.positions,
        pool.targets, pool.probas
    ))
    print(Counter(position_prediction))
```
