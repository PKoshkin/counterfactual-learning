```python
import numpy as np
import shutil
import os

from argparse import Namespace
from catboost import CatBoostRegressor, CatBoostClassifier

import sys
sys.path.append("../utils/")
sys.path.append("../calculating_predictions")
from calculate_predictions import calculate_predictions

%reload_ext autoreload
%autoreload 2
```

```python
def calculate_classification_targets_to_substruct(i, max_clicks):
    args = Namespace()
    args.verbose = True
    args.first_feature = 0
    args.last_feature = -1
    args.data_folder = "../../../data/best_features_days_data/"
    args.model_constructor = lambda verbose, max_clicks: CatBoostClassifier(
        verbose=False,
        loss_function='MultiClass',
        classes_count=max_clicks + 2
    )
    args.type = "classification"
    args.max_clicks = max_clicks
    args.need_position_feature = False
    out_folder = "classification_targets_to_substruct_max_clicks_" + str(max_clicks)
    args.train_days = [2]
    args.labels_to_substruct = None
    args.validation_day = None
    args.test_days = [3, 6]
    args.additional_features = {
        day: [
            os.path.join("features_models_no_pos", str(i), filename)
            for filename in os.listdir(os.path.join("features_models_no_pos", str(i)))
        ]
        for day in [2, 3, 6]
    }
    args.out_folder = os.path.join(out_folder, str(i))
    os.mkdir(args.out_folder)
    calculate_predictions(args)
```

```python
def calculate(max_clicks):
    out_folder = "classification_targets_to_substruct_max_clicks_" + str(max_clicks)
    if os.path.exists(out_folder):
        shutil.rmtree(out_folder)
    os.mkdir(out_folder)

    for i in range(15):
        calculate_classification_targets_to_substruct(i, max_clicks)
```

```python
%%time
for i in range(1, 6):
    calculate(i)
```
