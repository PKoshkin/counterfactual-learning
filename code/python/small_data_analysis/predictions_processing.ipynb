```python
def clear(folder):
    if os.path.exists(folder):
        shutil.rmtree(folder)
    os.mkdir(folder)
    
def get_metric(i, threshold):
    data_folder = "../../../data/best_features_days_data/"
    base_out_folder = "substructed_target_predictions_with_stacked_features"
    metric_folder = "metrics"

    out_folder = os.path.join(base_out_folder, str(i))

    targets_to_substruct = np.load(os.path.join(
        "classification_targets_to_substruct_max_clicks_5_edited",
        str(i),
        "train_2_test_6.npy"
    ))

    evaluate(
        out_folder,
        data_folder,
        metric_folder,
        argmax_positions,
        predictios_masks=[targets_to_substruct > threshold],
        verbose=False
    )

    with open("metrics/metrics.txt") as handler:
        metric = float(next(handler).strip().split()[0])
        return metric
```

```python
for threshold in np.linspace(0, 10, 100):
    metrics = [
        get_metric(i, threshold)
        for i in range(15)
    ]
    print(threshold, np.mean(metrics), np.std(metrics) / 15)
```
