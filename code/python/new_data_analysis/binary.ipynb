```python
import numpy as np
import os
import sys
from catboost import CatBoostClassifier
from collections import Counter
sys.path.append("../utils")
from pool_iterator import pool_iterator
from metric import calculate_metric
from json_tools import get_from_pool
from itertools import chain

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
train_days_filenames = [
    "../../../data/new_days_data_best_features/day_1.json",
    "../../../data/new_days_data_best_features/day_2.json",
    "../../../data/new_days_data_best_features/day_3.json"
]
test_day_filename = "../../../data/new_days_data_best_features/day_6.json"
position_variants = list(range(10)) + [100]
```

```python
%%time
train_features = np.concatenate([
    get_from_pool(pool_iterator(filename), "factors", np.array)
    for filename in train_days_filenames
    ], axis=0
)
train_targets = np.concatenate([
    get_from_pool(pool_iterator(filename), "target", lambda x: 1 if x > 0 else 0)
    for filename in train_days_filenames
    ], axis=0
)

test_features = get_from_pool(pool_iterator(test_day_filename), "factors", np.array)
test_targets = get_from_pool(pool_iterator(test_day_filename), "target", float)
test_probas = get_from_pool(pool_iterator(test_day_filename), "p", float)
test_positions = get_from_pool(pool_iterator(test_day_filename), "pos", int)
```

```python
model = CatBoostClassifier()
```

```python
%%time
model.fit(train_features, train_targets)
```

```python
predictions = model.predict(test_features).astype(int)
```

```python
answer = [4] * len(predictions)
metric = calculate_metric(
    answer, test_positions, test_targets, test_probas
)
print('', metric, Counter(answer))
```

```python
answer = np.array(position_variants)[np.random.randint(0, 11, size=len(answer))]
answer[predictions == 1] = 4
metric = calculate_metric(
    answer, test_positions, test_targets, test_probas
)
print('', metric, Counter(answer))
```

```python
best_positions = np.copy(test_positions)
best_positions[test_targets < 24] += 1
```

```python
metric = calculate_metric(
    best_positions, test_positions, test_targets, test_probas
)
print('', metric)
```
